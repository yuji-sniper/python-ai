```python
import numpy as np
import matplotlib.pyplot as plt

# 一次関数
x = np.linspace(-5, 5, 11)
y = 2*x + 1
plt.plot(x, y)
plt.show()
```

```python
# グラフの装飾
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 11)
y_1 = 1.5*x
y_2 = -2*x + 1

plt.xlabel("x value", size=14)
plt.ylabel("y value", size=14)

plt.title("My Graph", size=20)

plt.grid()

plt.plot(x, y_1, label="y_1")
plt.plot(x, y_2, label="y_2", linestyle="dashed")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4)
y_1 = 2*x + 1
y_2 = x**2 - 4
y_3 = 0.5*x**3 - 6*x

plt.plot(x, y_1, label="y_1")
plt.plot(x, y_2, label="y_2")
plt.plot(x, y_3, label="y_3")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2)

y_2 = 2**x
y_3 = 3**x

plt.plot(x, y_2, label="2^x")
plt.plot(x, y_3, label="3^x")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2)
e = np.e

y_2 = 2**x
y_e = e**x
y_3 = 3**x

plt.plot(x, y_2, label="2^x")
plt.plot(x, y_e, label="e^x")
plt.plot(x, y_3, label="3^x")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```
