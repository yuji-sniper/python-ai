```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()

iris_data = iris.data

st_data = iris_data[:50]
vc_data = iris_data[50:100]
vg_data = iris_data[100:150]

plt.scatter(st_data[:,0], st_data[:,1], label='setosa')
plt.scatter(vc_data[:,0], vc_data[:,1], label='versicolor')
plt.scatter(vg_data[:,0], vg_data[:,1], label='virginica')
plt.legend()
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()

plt.scatter(st_data[:,2], st_data[:,3], label='setosa')
plt.scatter(vc_data[:,2], vc_data[:,3], label='versicolor')
plt.scatter(vg_data[:,2], vg_data[:,3], label='virginica')
plt.legend()
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```
